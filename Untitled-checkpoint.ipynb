```python
!pip install mesa
import random
import mesa


class ConsumerAgent(mesa.Agent):
    def __init__(self, model):
        super().__init__(model)

        # Consumentvariabelen
        self.purchase_probability = random.uniform(0, 1)
        self.number_purchases = 0
        self.clothing_type = None  # 0 = fast-fashion, 1 = duurzaam

        self.environmental_awareness = random.uniform(0, 1)
        self.price_sensitivity = random.uniform(0, 1)
        self.fashion_sensitivity = random.uniform(0, 1)
        self.social_pressure_sensitivity = random.uniform(0, 1)

        self.recycling_probability = random.uniform(0, 1)
        self.number_recycled_items = 0

        self.disposal_probability = random.uniform(0, 1)
        self.number_disposed_items = 0

        self.discard_probability = random.uniform(0, 1)
        self.number_discarded_items = 0

        print(f"Consumer {self.unique_id} initialized")
        print(f"  Environmental_awareness: {self.environmental_awareness:.2f}")
        print(f"  Price_sensitivity: {self.price_sensitivity:.2f}")
        print(f"  Fashion_sensitivity: {self.fashion_sensitivity:.2f}")
        print(f"  Social_pressure_sensitivity: {self.social_pressure_sensitivity:.2f}")

    def step(self):
        # Reset aantallen per tijdstap
        self.number_purchases = 0
        self.number_recycled_items = 0
        self.number_disposed_items = 0
        self.number_discarded_items = 0

        # 1. AANKOOP VAN KLEDING
        self.purchase_probability = self.calculate_purchase_probability()

        if random.random() < self.purchase_probability:
            self.number_purchases = 1
            self.clothing_type = self.choose_clothing_type()

            if self.clothing_type == 0:
                self.model.total_fast_fashion_items += 1
                print(f"Consumer {self.unique_id} buys fast-fashion")
            else:
                self.model.total_sustainable_items += 1
                print(f"Consumer {self.unique_id} buys sustainable clothing")
        else:
            print(f"Consumer {self.unique_id} buys nothing this timestep")

        # 2. AFDANKEN VAN KLEDING
        self.recycling_probability = self.calculate_recycling_probability()

        if random.random() < self.recycling_probability:
            self.number_recycled_items = 1
            print(f"Consumer {self.unique_id} discards 1 clothing item")

            # 3. INLEVEREN VOOR RECYCLING OF WEGGOOIEN BIJ RESTAFVAL
            self.disposal_probability = self.calculate_disposal_probability()
            self.discard_probability = 1 - self.disposal_probability

            if random.random() < self.disposal_probability:
                self.number_disposed_items = 1
                self.model.total_recycled_textiles += 1
                print(f"Consumer {self.unique_id} hands in clothing for recycling")
            else:
                self.number_discarded_items = 1
                self.model.total_textile_waste += 1
                print(f"Consumer {self.unique_id} throws clothing in residual waste")
        else:
            print(f"Consumer {self.unique_id} does not discard clothing")

    def calculate_purchase_probability(self):
        # Invloed van mode, prijs, sociale druk en fast-fashion-heffing
        purchase_probability = (
            0.20
            + 0.30 * self.fashion_sensitivity
            + 0.20 * self.social_pressure_sensitivity
            - 0.25 * self.price_sensitivity * self.model.government.fast_fashion_tax
            - 0.15 * self.environmental_awareness
        )

        # Begrens tussen 0 en 1
        purchase_probability = max(0, min(1, purchase_probability))

        print(f"Consumer {self.unique_id} Purchase_probability: {purchase_probability:.2f}")
        return purchase_probability

    def choose_clothing_type(self):
        # 0 = fast-fashion, 1 = duurzaam
        fast_fashion_choice_probability = (
            0.50
            + 0.30 * self.fashion_sensitivity
            + 0.20 * self.price_sensitivity
            - self.model.effect_tax_purchase * self.model.government.fast_fashion_tax
            - 0.25 * self.environmental_awareness
        )

        # Begrens tussen 0 en 1
        fast_fashion_choice_probability = max(0, min(1, fast_fashion_choice_probability))

        print(
            f"Consumer {self.unique_id} fast-fashion choice probability: "
            f"{fast_fashion_choice_probability:.2f}"
        )

        if random.random() < fast_fashion_choice_probability:
            return 0
        else:
            return 1

    def calculate_recycling_probability(self):
        # Volgens jouw tabel:
        # Recycling_probability = kans dat een consument een kledingstuk afdankt
        recycling_probability = (
            0.10
            + 0.20 * self.fashion_sensitivity
            + 0.10 * self.social_pressure_sensitivity
        )

        recycling_probability = max(0, min(1, recycling_probability))

        print(f"Consumer {self.unique_id} Recycling_probability: {recycling_probability:.2f}")
        return recycling_probability

    def calculate_disposal_probability(self):
        # Disposal_probability = kans op inleveren voor recycling
        disposal_probability = (
            0.20
            + 0.35 * self.environmental_awareness
            + self.model.effect_reward_recycling * self.model.government.reward
        )

        disposal_probability = max(0, min(1, disposal_probability))

        print(f"Consumer {self.unique_id} Disposal_probability: {disposal_probability:.2f}")
        return disposal_probability


class GovernmentAgent(mesa.Agent):
    def __init__(self, model):
        super().__init__(model)

        # Overheidvariabelen
        self.reward = model.start_reward
        self.fast_fashion_tax = model.start_fast_fashion_tax

        print("Government initialized")
        print(f"  Reward: {self.reward}")
        print(f"  Fast_fashion_tax: {self.fast_fashion_tax}")

    def step(self):
        # In dit model blijft beleid constant binnen een scenario
        print("Government step")
        print(f"  Reward: {self.reward}")
        print(f"  Fast_fashion_tax: {self.fast_fashion_tax}")


class TextileWasteModel(mesa.Model):
    def __init__(self, number_consumers, start_reward, start_fast_fashion_tax, seed=None):
        super().__init__(seed=seed)

        # Modelvariabelen
        self.number_consumers = number_consumers
        self.effect_reward_recycling = 0.20
        self.effect_tax_purchase = 0.20

        self.total_recycled_textiles = 0
        self.total_textile_waste = 0
        self.total_fast_fashion_items = 0
        self.total_sustainable_items = 0

        self.start_reward = start_reward
        self.start_fast_fashion_tax = start_fast_fashion_tax

        # Maak overheid aan
        self.government = GovernmentAgent(self)

        # Maak consumenten aan
        self.consumers = [ConsumerAgent(self) for _ in range(self.number_consumers)]

        print("Model initialized")
        print(f"  Number_consumers: {self.number_consumers}")
        print(f"  Effect_reward_recycling: {self.effect_reward_recycling}")
        print(f"  Effect_tax_purchase: {self.effect_tax_purchase}")

    def step(self):
        # Reset modeluitkomsten per tijdstap
        self.total_recycled_textiles = 0
        self.total_textile_waste = 0
        self.total_fast_fashion_items = 0
        self.total_sustainable_items = 0

        self.government.step()

        # Laat alle consumenten een stap doen
        random.shuffle(self.consumers)
        for consumer in self.consumers:
            consumer.step()

        print("Model step completed")
        print(f"  Total_recycled_textiles: {self.total_recycled_textiles}")
        print(f"  Total_textile_waste: {self.total_textile_waste}")
        print(f"  Total_fast_fashion_items: {self.total_fast_fashion_items}")
        print(f"  Total_sustainable_items: {self.total_sustainable_items}")


# =========================
# VOORBEELD VAN GEBRUIK
# =========================
if __name__ == "__main__":
    model = TextileWasteModel(
        number_consumers=1000,
        start_reward=0,
        start_fast_fashion_tax=0
    )

    for t in range(3):
        print(f"\n--- TIMESTEP {t + 1} ---")
        model.step()
        
```
